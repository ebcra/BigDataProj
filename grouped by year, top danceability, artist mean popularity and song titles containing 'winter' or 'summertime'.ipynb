```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline 
```

```python
'''Spotify's description of Danceabillity: Danceability describes how suitable a track is for dancing based on a combination of musical elements including
tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most 
danceable'''
```

```python
'''Spotify's description of popularity: The popularity of the track. The value will be between 0 and 100, with 100 being the most popular.
The popularity of a track is a value between 0 and 100, with 100 being the most popular. The popularity is 
calculated by algorithm and is based, in the most part, on the total number of plays the track has had and how 
recent those plays are.
Generally speaking, songs that are being played a lot now will have a higher popularity than songs that were 
played a lot in the past. Duplicate tracks (e.g. the same track from a single and an album) are rated 
independently. Artist and album popularity is derived mathematically from track popularity. Note that the 
popularity value may lag actual popularity by a few days: the value is not updated in real time.'''
```

```python
#reading csv with the datasaet of 160k spotify songs
df = pd.read_csv('data.csv')
```

```python
#50 most danceable tracks
top50dance = df.danceability.nlargest(50)
```

```python
#rounding to 3 decimals
top50dance = round(top50dance,3)
```

```python
#getting mean danceability among top 50 most danceable tracks
df.danceability.mean()
```

```python
#the top 10 % of songs with the highest danceability
dancetop10 = df.nlargest(16991, 'danceability')
dancetop10.describe()
```

```python
#d=top10% of songs with highest danceability
#p=top10% of songs with highest popularity
#didnt choose all columns
d = df[['name', 'artists','acousticness', 'danceability', 'energy',
    'instrumentalness', 'liveness', 'popularity', 'speechiness',
       'valence']].nlargest(16991, 'danceability')
p = df[['name', 'artists','acousticness', 'danceability', 'energy',
    'instrumentalness', 'liveness', 'popularity', 'speechiness',
       'valence']].nlargest(16991, 'popularity')
```

```python
df.columns
```

```python
#joining p & d to see which songs that are both among top10% danceability and popularity. counting them
pandd = p.merge(d, 'inner')
pandd.name.count()
```

```python
pandd
```

```python
# artists and number of songs they have in top10%dance and popularity. and their mean dance and popu
topdance_pop=pandd.groupby('artists').agg({'name':'count', 'popularity':'mean', 'danceability':'mean'}).sort_values(by=['name', 'popularity', 'danceability'], ascending=[False,False,False])
topdance_pop['tracks'] = topdance_pop.name
topdance_pop.drop(columns='name', inplace=True)
topdance_pop.head(20)
```

```python
#creating xsticks fot x-axis names
xsticks = ['Lil Uzi Vert', 'Drake', 'Eminem', '$uicideBoy$',
'Michael Jackson', 'Bob Marley & The Wailers', 'DaBaby',
'Future', 'Bad Bunny', 'YoungBoy Never Broke Again',
'XXXTENTACION', 'J. Cole', 'Lil Baby', 'Gunna',
'Mac Miller', 'Aventura', 'Polo G',
'Banda MS de Sergio Lizárraga', 'Khalid',
'A Boogie Wit da Hoodie']
```

```python
#these artists have the most songs among both top10% dance and popular. the mean dace and popularity of those
#songs is displayed as well as no of songs 

topdance_pop.head(20).plot(kind='line', subplots=True, figsize=(15,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.xticks(range(20), xsticks)
plt.show()
```

```python
#grouping by year and looking at mean of all parameters for each year
year = df.groupby(['year']).agg({'name':'count', 'acousticness':'mean', 'danceability':'mean', 'energy':'mean', 
                                 'explicit':'mean','instrumentalness':'mean', 'key':'mean', 'liveness':'mean', 
                                 'loudness':'mean','popularity':'mean', 'speechiness':'mean', 'tempo':'mean', 
                                 'valence':'mean'})
```

```python
#sorting by year
year = year.sort_values(by='year')
```

```python
# getting range index to have year as column
year.reset_index(inplace=True)
```

```python
#looking at correlations
year.corr()
```

```python
#wanting to see those correlations as graphs. in this case acousticness and year
acousticness = year[['acousticness','year']].plot(kind='line', x='year', y='acousticness', figsize=(12,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Year', fontsize=15, c='lime')
plt.ylabel('Average acousticness', fontsize=15, c='lime')
plt.title('Average acousicness for each year', fontsize=20, c='lime')
plt.show()
```

```python
#and in this case whether songs are explicit or not and year
year.reset_index()[['explicit','year']].plot(kind='line', x='year', y='explicit', figsize=(12,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Year', fontsize=15, c='lime')
plt.ylabel("Percentage of songs labelled 'Explicit'", fontsize=15, c='lime')
plt.title('Explicit over the years', fontsize=20, c='lime')
```

```python
#grouping by artist and getting artist mean popularity and number of tracks in set
songs_artist = df.groupby('artists').agg({'popularity':'mean', 'name':'count'})
```

```python
#artists with more than 2 songs. number of songs and mean popularity
#removed all with just one song as they might be collabs/featurings or onehitwonders

artist_mean_pop = songs_artist.loc[songs_artist.name > 2].nlargest(100, 'popularity')
```

```python
#ordered by artists mean popularity and secondly their number of songs
artist_mean_pop.sort_values(by=['popularity', 'name'], ascending=[False, False]).head(20).plot(kind='bar', figsize=(12,8), color=['g', 'lime'] , fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel("Number of songs and artist mean popularity", fontsize=15, c='lime')
plt.title('Top 20 artists with highest popularity mean', fontsize=20, c='lime')
plt.show()
#only arists with more than two songs in set.  
```

```python
#ordered by how many songs the artist has and secondly by their mean popularity
artist_mean_pop_nsongs = songs_artist.sort_values(by=['name','popularity'] , ascending=[False,False]).head(20)
artist_mean_pop_nsongs.plot(kind='bar', figsize=(12,8), color=['g', 'lime'] , fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel("Number of songs and artist mean popularity", fontsize=15, c='lime')
plt.title('Top 20 artists with highest popularity mean', fontsize=20, c='lime')
plt.show()
```
