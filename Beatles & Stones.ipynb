```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline 
```

```python
#reading csv
df = pd.read_csv('data.csv')
```

```python
#checking which artist(s) have the most tracks in the set

df['artists'].value_counts().head(15)
```

```python
#which 15 artists have the most songs in the set
df['artists'].value_counts().head(15).plot(kind='bar', figsize=(12,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel('No of Tracks', fontsize=15, c='lime')
plt.title('Top 15 artists with most songs', fontsize=20, c='lime')
plt.show()
```

```python
#how amny songs do beatles and stones have?
df.loc[(df.artists== "['The Beatles']") | (df.artists == "['The Rolling Stones']")].groupby('artists').id.count().plot(kind='bar', figsize=(12,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel('No of Tracks', fontsize=15, c='lime')
plt.title('Beatles & Stones no of songs', fontsize=20, c='lime')
plt.show()
```

```python
#choosing all beatles songs
beatles = df.loc[df.artists == "['The Beatles']"]
```

```python
#choosing all stones songs
stones = df.loc[df.artists == "['The Rolling Stones']"]
```

```python
#looking at beatles 30 most popular songs
beatles_top30=beatles.loc[beatles.popularity >= 65].sort_values(by='popularity', ascending=False).head(30)
beatles_top30
```

```python
#what precentage of beatles songs have a popularity of 75 or higher
beatles.loc[beatles.popularity >= 75][['name']].count()/beatles.name.count()
```

```python
#looking at stones 30 most popular songs 
stones_top30=stones.loc[stones.popularity >= 50].sort_values(by='popularity', ascending=False).head(30)
stones_top30
```

```python
#what precentage of stones songs have a popularity of 75 or higher
stones.loc[stones.popularity >= 75][['name']].count() / stones.name.count()
```

```python
#mean popularity of beatles songs
beatles.popularity.mean()
```

```python
#mean popularity of stones songs
stones.popularity.mean()
```

```python
#showing beatles and stones songs mean popularity
beatles.merge(stones, how='outer').groupby('artists').popularity.mean().plot(kind='bar', figsize=(12,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel('Mean popularity', fontsize=15, c='lime')
plt.title('Beatles & Stones mean popularity', fontsize=20, c='lime')

```

```python
#merging beatles and stones songs and looking at their reallt popular songs
sb=beatles.merge(stones, 'outer')
sb.loc[sb.popularity>=75].groupby('artists').id.count().plot(kind='bar', figsize=(12,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel('No of songs with popularity of at least 75', fontsize=15, c='lime')
plt.title('Beatles & Stones really popular', fontsize=20, c='lime')

##the beatles have 0.014634 of songs with popularity of 75 or more
##the stones have 0.0.009766 of songs with popularity of 75 or more
```

```python
# betles 30 least popular songs
beatles_bottom30 = beatles.loc[beatles.popularity <= 30].sort_values(by = 'popularity').head(30)
beatles_bottom30
```

```python
#what precentage of beatles songs have a pop of 30 or lower
beatles.loc[beatles.popularity <= 30].sort_values(by = 'popularity')[['name']].count()/ beatles.name.count()
```

```python
# stones 30 least popular songs
stones_bottom30 = stones.loc[stones.popularity <= 30].sort_values(by = 'popularity').head(30)
stones_bottom30
```

```python
#what precentage of stones songs have a pop of 30 or lower
stones.loc[stones.popularity < 30].sort_values(by = 'popularity')[['name']].count()/ stones.name.count()
```

```python
#how many songs do beatles and stones have with popularity of less than 30
sb.loc[sb.popularity<30].groupby('artists').id.count()
```

```python
#looking at beatles and stones songs with popularity of 30 or less
sb.loc[sb.popularity<30].groupby('artists').id.count().plot(kind='bar', figsize=(12,8), color='lime', fontsize=12)
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel('No of songs with popularity of 30 or less', fontsize=15, c='lime')
plt.title('Beatles & Stones least popular', fontsize=20, c='lime')

##the beatles have 0.12439 of songs with popularity of 30 or less
##the stones have 0.402344of songs with popularity of 30 or less
```

```python
#looking at bealtes and stones songs averages on differnt parameters
sb.groupby('artists')[['acousticness', 'danceability', 'energy', 'explicit',
       'instrumentalness',  'liveness',  
        'speechiness',  'valence']].mean().plot(kind='bar', figsize=(12,8), fontsize=12, colormap='YlGn')
plt.style.use('dark_background')
plt.xlabel('Artist', fontsize=15, c='lime')
plt.ylabel('Average for all songs', fontsize=15, c='lime')
plt.title('Beatles & Stones averages', fontsize=20, c='lime')
```
